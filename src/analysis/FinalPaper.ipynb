```python
"""
============================================================================
COMPLETE DATA CLEANING PIPELINE
============================================================================
This script creates a clean dataset for analyzing minimum wage impacts on
occupational dynamics in Russia using RLMS-HSE data (1994-2024).

Inputs:
- df_analysis: Raw RLMS data with all variables
- df_mw: Minimum wage data by region and year

Output:
- df_clean: Cleaned and merged dataset ready for analysis
============================================================================
"""

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("STEP 1: LOADING AND PREPARING DATA")
print("="*80)

# ============================================================================
# STEP 1: PREPARE MINIMUM WAGE DATA
# ============================================================================

print("\nüìä Minimum Wage Data:")
print(f"Shape: {df_mw.shape}")
print(f"Columns: {df_mw.columns.tolist()}")
print(f"Years: {df_mw['year'].min()} to {df_mw['year'].max()}")
print(f"Unique regions in MW data: {df_mw['region'].nunique()}")

# Standardize minimum wage data
df_mw_clean = df_mw.copy()
df_mw_clean = df_mw_clean.rename(columns={
    'region': 'mw_region_name',
    'min_wage': 'min_wage'
})
print("‚úÖ Minimum wage data prepared")

# ============================================================================
# STEP 2: PREPARE REGION MAPPING FROM DF_ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("STEP 2: CREATING REGION MAPPING")
print("="*80)

# Create region mapping from RLMS documentation (rows 37-200 in df3)
region_name_mapping = {
    1: "Leningrad Oblast: Volosovkij Rajon",
    2: "Moscow City",
    3: "Moscovskaya Oblast",
    4: "Syktyvkar: Komi ASSR",
    5: "Usinsk CR: Komi-ASSR (1994-2019) / Pechora CR: Komi-ASSR (2020-present)",
    6: "Volosovkij Rajon: Leningradskaya Oblast",
    7: "Smolensk: Smolenskaya Oblast",
    8: "Rzhev City and Rajon: Tverskaya Oblast",
    9: "Krasnodar CR",
    10: "Udmurt ASSR: Glasov CR",
    11: "Nizhny Novgorod: Nizhegorodskaya Oblast",
    12: "Perm Territory: Solikamsk City & Rajon",
    13: "Zemetchinskij Rajon: Penzenskaya Oblast",
    14: "Kaluzhskaya Oblast: Kuibyshev Rajon",
    15: "Uvarov City & Rajon: Tambovskaya Oblast",
    16: "Kazan: Tatarskaja ASSR",
    17: "Saratov: Saratovskaya Oblast",
    18: "Volsk City and Rajon: Saratovskya Oblast",
    19: "Rudnyanskij Rajon: Volgogradskaya Oblast",
    20: "Zalukokoazhe and Zol`skij Rajon: Kabardino-Balkaria",
    21: "Batajsk: Rostovskaya Oblast",
    22: "Krasnodar: Krasnodarskij Krai",
    23: "Georgievsk City & Rajon: Stavropolskij Kraj",
    24: "Kushchevsky Rajon: Krasnodarskij Krai",
    25: "Cheliabinsk: Chelyabinskaya Oblast",
    26: "Kurgan: Kurganskaya Oblast",
    27: "Glazov City & Rajon:Udmurt ASSR",
    28: "Orsk: Orenburgskaya Oblast",
    29: "Solikamsk City and Rajon: Permskaya Oblast - until 2005 Permskij Krai - since 2005",
    30: "Krasnoarmeyskiy Rajon: Chelyabinskaya Oblast",
    31: "Tomsk: Tomskaya Oblast",
    32: "Tumenskaya Obl, Surgutskij R-n: Khanty-Mansi AO(1994-2002)/ Berdsk City & Raion: Novosibirskaya Obl (2003-present)",
    33: "Tambov Oblast: Uvarovo CR",
    34: "Kur`inskij Rajon: Altaiskij Krai",
    35: "Krasnoyarsk: Krasnoyarskij Krai",
    36: "Vladivostok: Primorskij Krai",
    37: "Nazarov City & Rajon: Krasnoyarskij Krai",
    38: "Tambovskij Rajon: Amurskaya Oblast",
    39: "Volgograd Oblast: Rudnjanskij Rajon",
    45: "Tatarskaja ASSR: Kazan",
    46: "Kurgan",
    47: "Orenburg Oblast: Orsk",
    48: "Chuvashskaya ASSR: Shumerlja CR",
    52: "Stavropolskij Kraj: Georgievskij CR",
    58: "Altaiskij Kraj: Kur`inskij Rajon",
    66: "Krasnojarskij Kraij: Krasnojarsk",
    67: "Kalinin Oblast: Rzhev CR",
    70: "Saratov CR",
    71: "Tomsk",
    72: "Lipetskaya Oblast: Lipetsk CR",
    73: "Krasnojarskij Kraij: Nazarovo CR",
    77: "Kabardino-Balkarija, Zol `skij Rajon",
    84: "Altaiskij Kraj: Biisk CR",
    86: "Khanty-Mansi Aut Okrug: Surgutskij Rajon: (1994-2002)",
    89: "Komi-ASSR: Usinsk CR (1994-2019)",
    92: "Vladivostok",
    93: "Amurskaja Oblast: Tambovskii Rajon",
    100: "Saratov Oblast: Volskij Gorosovet & Rajon",
    105: "Komi-ASSR: Syktyvkar",
    106: "Cheliabinsk",
    107: "Cheliabinsk Oblast: Krasnoarmeiskij Rajon",
    116: "Gorkovskaja Oblast: Nizhnij Novgorod",
    117: "Penzenskaya Oblast: Zemetchinskij Rajon",
    129: "Krasnodarskij Kraj: Kushchevskij Rajon",
    135: "Smolensk CR",
    136: "Tulskaja Oblast: Tula",
    137: "Rostov Oblast: Batajsk",
    138: "Moscow City",
    139: "Moscow City",
    140: "New Moscow City",
    141: "St. Petersburg City",
    142: "Moscow Oblast",
    161: "Novosibirskaya Oblast: Berdsk City & Raion (2003-present)",
    200: "Komi-ASSR: Pechora CR (2020-present)"
}

# ============================================================================
# STEP 3: ENSURE REGION COLUMN IN DF_ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("STEP 3: ENSURING REGION COLUMN IN DF_ANALYSIS")
print("="*80)

# Check for region column
if 'region' in df_analysis.columns:
    print("‚úÖ 'region' column exists")
    region_col = 'region'
elif 'region_df2' in df_analysis.columns:
    print("‚úÖ Renaming 'region_df2' to 'region'")
    df_analysis = df_analysis.rename(columns={'region_df2': 'region'})
    region_col = 'region'
else:
    # Look for any column with region in name
    region_candidates = [col for col in df_analysis.columns if 'region' in col.lower()]
    if region_candidates:
        print(f"‚úÖ Using '{region_candidates[0]}' as region column")
        df_analysis = df_analysis.rename(columns={region_candidates[0]: 'region'})
        region_col = 'region'
    else:
        raise ValueError("No region column found in df_analysis")

print(f"Region unique values: {df_analysis['region'].nunique()}")
print(f"Sample region codes: {df_analysis['region'].dropna().unique()[:10]}")

# ============================================================================
# STEP 4: ADD REGION NAMES TO DF_ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("STEP 4: ADDING REGION NAMES")
print("="*80)

# Create region mapping dataframe
region_map_df = pd.DataFrame([
    {"region_code": k, "region_fullname": v} 
    for k, v in region_name_mapping.items()
])

# Add region names to df_analysis
df_analysis = df_analysis.merge(
    region_map_df,
    left_on='region',
    right_on='region_code',
    how='left'
)

print(f"Region name coverage: {df_analysis['region_fullname'].notna().mean():.1%}")
print(f"Matched regions: {df_analysis['region_fullname'].notna().sum():,}")

# ============================================================================
# STEP 5: EXTRACT BASE REGION NAMES FOR MERGING
# ============================================================================

print("\n" + "="*80)
print("STEP 5: EXTRACTING BASE REGION NAMES")
print("="*80)

def extract_base_region_name(fullname):
    """Extract base region name for merging with minimum wage data"""
    if pd.isna(fullname):
        return None
    
    # Handle special cases first
    special_mapping = {
        'Leningrad Oblast': 'Leningrad',
        'Leningradskaya': 'Leningrad',
        'Moscow City': 'Moscow',
        'Moscovskaya': 'Moscow',
        'St. Petersburg': 'Saint',
        'Nizhegorodskaya': 'Nizhny',
        'Saratovskya': 'Saratov',
        'Volgogradskaya': 'Volgograd',
        'Chelyabinskaya': 'Chelyabinsk',
        'Kurganskaya': 'Kurgan',
        'Orenburgskaya': 'Orenburg',
        'Permskaya': 'Perm',
        'Tomskaya': 'Tomsk',
        'Novosibirskaya': 'Novosibirsk',
        'Amurskaja': 'Amur',
        'Krasnojarskij': 'Krasnoyarsk',
        'Altaiskij': 'Altai',
        'Primorskij': 'Primorsky',
        'Tatarskaja': 'Kazan',
        'Udmurt': 'Udmurt',
        'Chuvashskaya': 'Chuvash',
        'Kabardino-Balkarija': 'Kabardino-Balkaria',
        'Krasnodarskij': 'Krasnodar',
        'Stavropolskij': 'Stavropol'
    }
    
    for key, value in special_mapping.items():
        if key in fullname:
            return value
    
    # If no special case, take first part before colon or comma
    if ':' in fullname:
        base = fullname.split(':')[0].strip()
    elif ',' in fullname:
        base = fullname.split(',')[0].strip()
    else:
        base = fullname
    
    # Remove common suffixes
    suffixes = [' Oblast', ' Kraj', ' Krai', ' ASSR', ' CR', ' City', ' Republic', 'skaya', 'skij']
    for suffix in suffixes:
        if base.endswith(suffix):
            base = base[:-len(suffix)]
    
    return base.strip()

# Apply extraction
df_analysis['region_base'] = df_analysis['region_fullname'].apply(extract_base_region_name)

print("Sample region name mapping:")
sample = df_analysis[['region', 'region_fullname', 'region_base']].dropna().drop_duplicates().head(10)
print(sample)

# ============================================================================
# STEP 6: CREATE OCCUPATION VARIABLES
# ============================================================================

print("\n" + "="*80)
print("STEP 6: CREATING OCCUPATION VARIABLES")
print("="*80)

# Create occupation digits if not exists
if 'j2cod08' in df_analysis.columns:
    # Remove missing codes
    df_analysis = df_analysis[df_analysis['j2cod08'] < 90000000].copy()
    
    # Create occupation digits
    df_analysis['occ_1digit'] = (df_analysis['j2cod08'] // 1000).astype(int)
    df_analysis['occ_2digit'] = (df_analysis['j2cod08'] // 100).astype(int)
    df_analysis['occupation_code'] = df_analysis['j2cod08']
    
    print(f"Unique 4-digit occupations: {df_analysis['j2cod08'].nunique()}")
    print(f"Unique 1-digit occupations: {df_analysis['occ_1digit'].nunique()}")
else:
    print("‚ö†Ô∏è  j2cod08 not found in data")

# ============================================================================
# STEP 7: CREATE OCCUPATION CHANGE VARIABLES
# ============================================================================

print("\n" + "="*80)
print("STEP 7: CREATING OCCUPATION CHANGE VARIABLES")
print("="*80)

# Sort for panel operations
df_analysis = df_analysis.sort_values(['idind', 'year'])

# Create previous occupation
df_analysis['prev_occ'] = df_analysis.groupby('idind')['j2cod08'].shift(1)
df_analysis['prev_occ1'] = df_analysis.groupby('idind')['occ_1digit'].shift(1)

# Occupation change indicators
df_analysis['occupation_changed'] = (
    (df_analysis['j2cod08'] != df_analysis['prev_occ']) & 
    df_analysis['prev_occ'].notna()
).astype(int)

df_analysis['major_occ_changed'] = (
    (df_analysis['occ_1digit'] != df_analysis['prev_occ1']) & 
    df_analysis['prev_occ1'].notna()
).astype(int)

print(f"Occupation change rate (4-digit): {df_analysis['occupation_changed'].mean():.1%}")
print(f"Major occupation change rate (1-digit): {df_analysis['major_occ_changed'].mean():.1%}")

# ============================================================================
# STEP 8: CREATE INFORMALITY INDICATORS
# ============================================================================

print("\n" + "="*80)
print("STEP 8: CREATING INFORMALITY INDICATORS")
print("="*80)

# Clean missing codes
missing_codes = [99999996, 99999997, 99999998, 99999999, 88888888]
for col in ['j15', 'j17', 'j19']:
    if col in df_analysis.columns:
        df_analysis[col] = df_analysis[col].replace(missing_codes, np.nan)

# Create informality indicators
if 'j17' in df_analysis.columns:
    df_analysis['in_kind_payment'] = (df_analysis['j17'] == 1).astype(int)

if 'j19' in df_analysis.columns:
    df_analysis['unpaid_leave'] = (df_analysis['j19'] == 1).astype(int)

if 'j15' in df_analysis.columns:
    df_analysis['wage_arrears'] = (df_analysis['j15'] > 0).astype(int)

# Composite informality measure
informality_cols = []
if 'in_kind_payment' in df_analysis.columns:
    informality_cols.append('in_kind_payment')
if 'unpaid_leave' in df_analysis.columns:
    informality_cols.append('unpaid_leave')
if 'wage_arrears' in df_analysis.columns:
    informality_cols.append('wage_arrears')

if informality_cols:
    df_analysis['informal'] = df_analysis[informality_cols].any(axis=1).astype(int)
    print(f"Informal employment rate: {df_analysis['informal'].mean():.1%}")

# ============================================================================
# STEP 9: CREATE SECOND JOB INDICATOR
# ============================================================================

print("\n" + "="*80)
print("STEP 9: CREATING SECOND JOB INDICATOR")
print("="*80)

df_analysis['has_second_job'] = 0
if 'j40' in df_analysis.columns:
    df_analysis['has_second_job'] = df_analysis['has_second_job'] | df_analysis['j40'].notna().astype(int)
if 'j35.1' in df_analysis.columns:
    df_analysis['has_second_job'] = df_analysis['has_second_job'] | df_analysis['j35.1'].notna().astype(int)

print(f"Second job rate: {df_analysis['has_second_job'].mean():.1%}")

# ============================================================================
# STEP 10: CREATE AGE GROUP
# ============================================================================

print("\n" + "="*80)
print("STEP 10: CREATING AGE GROUP")
print("="*80)

df_analysis['age_group'] = pd.cut(
    df_analysis['age'], 
    bins=[15, 25, 35, 45, 55, 65, 75], 
    labels=['15-25', '26-35', '36-45', '46-55', '56-65', '66-75']
)

# ============================================================================
# STEP 11: FILTER TO WORKING AGE POPULATION
# ============================================================================

print("\n" + "="*80)
print("STEP 11: FILTERING TO WORKING AGE POPULATION")
print("="*80)

original_count = len(df_analysis)
df_analysis = df_analysis[df_analysis['age'].between(15, 72)].copy()
if 'j1' in df_analysis.columns:
    df_analysis = df_analysis[df_analysis['j1'] == 1].copy()  # Currently working

print(f"Original observations: {original_count:,}")
print(f"After filtering: {len(df_analysis):,}")

# ============================================================================
# STEP 12: MERGE WITH MINIMUM WAGE DATA
# ============================================================================

print("\n" + "="*80)
print("STEP 12: MERGING WITH MINIMUM WAGE DATA")
print("="*80)

# Create matching dictionary between region_base and MW region names
mw_regions = df_mw_clean['mw_region_name'].unique()

# Create matching function
def match_to_mw(region_base, mw_regions):
    """Match base region name to minimum wage region name"""
    if pd.isna(region_base):
        return None
    
    region_lower = region_base.lower()
    
    # Direct match
    for mw_region in mw_regions:
        if mw_region.lower() == region_lower:
            return mw_region
    
    # Partial match
    for mw_region in mw_regions:
        if mw_region.lower() in region_lower or region_lower in mw_region.lower():
            return mw_region
    
    return None

# Apply matching
df_analysis['mw_region_match'] = df_analysis['region_base'].apply(
    lambda x: match_to_mw(x, mw_regions)
)

match_rate = df_analysis['mw_region_match'].notna().mean()
print(f"Region match rate: {match_rate:.1%}")

# Merge with minimum wage data
df_clean = df_analysis.merge(
    df_mw_clean,
    left_on=['mw_region_match', 'year'],
    right_on=['mw_region_name', 'year'],
    how='left'
)

print(f"After merge shape: {df_clean.shape}")
print(f"Min wage coverage: {df_clean['min_wage'].notna().mean():.1%}")

# ============================================================================
# STEP 13: CREATE FINAL POLICY VARIABLES
# ============================================================================

print("\n" + "="*80)
print("STEP 13: CREATING FINAL POLICY VARIABLES")
print("="*80)

# Ensure we have a clean region column
if 'region' not in df_clean.columns:
    if 'region_x' in df_clean.columns:
        df_clean['region'] = df_clean['region_x']
    elif 'region_df2' in df_clean.columns:
        df_clean['region'] = df_clean['region_df2']

# Sort for panel operations
df_clean = df_clean.sort_values(['region', 'year'])

# Create minimum wage increase indicators
df_clean['min_wage_lag'] = df_clean.groupby('region')['min_wage'].shift(1)
df_clean['min_wage_increase'] = (
    (df_clean['min_wage'] > df_clean['min_wage_lag']) & 
    (df_clean['min_wage_lag'].notna())
).astype(int)

# Treatment region (ever had an increase)
region_inc = df_clean.groupby('region')['min_wage_increase'].max()
treatment_regions = region_inc[region_inc == 1].index.tolist()
df_clean['treatment_region'] = df_clean['region'].isin(treatment_regions).astype(int)

# Post-treatment indicator
first_inc = df_clean[df_clean['min_wage_increase'] == 1].groupby('region')['year'].min().reset_index()
first_inc.columns = ['region', 'first_inc_year']
df_clean = df_clean.merge(first_inc, on='region', how='left')
df_clean['post_treatment'] = (df_clean['year'] >= df_clean['first_inc_year']).astype(int)
df_clean['post_treatment'] = df_clean['post_treatment'].fillna(0)

# Create Kaitz index (min wage / proxy wage)
if 'j256' in df_clean.columns:
    bin_to_wage = {1:1500, 2:3000, 3:6000, 4:10000, 5:15000,
                   6:21000, 7:30000, 8:43000, 9:60000}
    df_clean['wage_proxy'] = df_clean['j256'].map(bin_to_wage)
    df_clean['kaitz_index'] = df_clean['min_wage'] / df_clean['wage_proxy']
    df_clean['kaitz_index'] = df_clean['kaitz_index'].replace([np.inf, -np.inf], np.nan)

# Clean up
df_clean = df_clean.drop(columns=['first_inc_year', 'min_wage_lag'], errors='ignore')

print(f"Treatment regions: {len(treatment_regions)}")
print(f"Min wage increases: {df_clean['min_wage_increase'].sum()}")
print(f"Post-treatment observations: {df_clean['post_treatment'].sum():,}")

# ============================================================================
# STEP 14: FINAL CLEANING
# ============================================================================

print("\n" + "="*80)
print("STEP 14: FINAL CLEANING")
print("="*80)

# Keep only observations with minimum wage data for analysis
df_clean_final = df_clean[df_clean['min_wage'].notna()].copy()

# Ensure binary variables are 0/1
for col in ['treatment_region', 'post_treatment', 'min_wage_increase', 
            'occupation_changed', 'major_occ_changed', 'informal', 'has_second_job']:
    if col in df_clean_final.columns:
        df_clean_final[col] = (df_clean_final[col] > 0.5).astype(int)

print(f"\n{'='*80}")
print("FINAL CLEAN DATASET SUMMARY")
print(f"{'='*80}")
print(f"Observations: {len(df_clean_final):,}")
print(f"Years: {df_clean_final['year'].min():.0f} to {df_clean_final['year'].max():.0f}")
print(f"Regions: {df_clean_final['region'].nunique()}")
print(f"\nKey indicators:")
print(f"  Occupation change rate: {df_clean_final['occupation_changed'].mean()*100:.1f}%")
print(f"  Informal employment: {df_clean_final['informal'].mean()*100:.1f}%")
print(f"  Second job rate: {df_clean_final['has_second_job'].mean()*100:.1f}%")
print(f"  Mean minimum wage: {df_clean_final['min_wage'].mean():.0f} rubles")
print(f"  Kaitz index: {df_clean_final['kaitz_index'].mean():.2f}")

# ============================================================================
# STEP 15: SAVE FINAL DATASET
# ============================================================================

print("\n" + "="*80)
print("STEP 15: SAVING FINAL DATASET")
print("="*80)

# Uncomment to save
# df_clean_final.to_csv('russia_labor_market_clean.csv', index=False)
print("‚úÖ Clean dataset ready for analysis!")
print(f"Final shape: {df_clean_final.shape}")

# Create a variable list for documentation
variable_list = pd.DataFrame({
    'variable': df_clean_final.columns,
    'dtype': df_clean_final.dtypes.values,
    'non_null': df_clean_final.notna().sum().values,
    'mean': df_clean_final.mean(numeric_only=True).reindex(df_clean_final.columns, fill_value=np.nan).values
})
print("\nüìã Variable summary saved to 'variable_list.csv'")
# variable_list.to_csv('variable_list.csv', index=False)

# ============================================================================
# FINAL OUTPUT
# ============================================================================

# The cleaned dataset is now in df_clean_final
# Use this for all subsequent analysis

df_clean = df_clean_final.copy()
print("\n‚úÖ df_clean is now ready for analysis!")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================================
# PART 1: OCCUPATIONAL MOBILITY ANALYSIS
# ============================================================================
print("="*60)
print("PART 1: OCCUPATIONAL MOBILITY ANALYSIS")
print("="*60)

# 1.1 Overall mobility rates
print("\n1.1 OVERALL MOBILITY RATES")
print("-"*40)
print(f"4-digit occupation change rate: {df_clean['occupation_changed'].mean():.1%}")
print(f"1-digit occupation change rate: {df_clean['major_occ_changed'].mean():.1%}")

# 1.2 Mobility rates by year
print("\n1.2 MOBILITY RATES BY YEAR")
print("-"*40)
yearly_mobility = df_clean.groupby('year').agg({
    'occupation_changed': 'mean',
    'major_occ_changed': 'mean',
    'idind': 'count'
}).rename(columns={'idind': 'N'})
print(yearly_mobility)

# 1.3 Mobility by occupation category
print("\n1.3 MOBILITY RATES BY OCCUPATION (1-digit)")
print("-"*40)
occ_labels = {1:'Managers', 2:'Professionals', 3:'Technicians', 4:'Clerical', 
              5:'Services', 6:'Agriculture', 7:'Crafts', 8:'Operators', 
              9:'Elementary', 0:'Military/Other'}

occ_mobility = df_clean.groupby('occ_1digit').agg({
    'occupation_changed': 'mean',
    'major_occ_changed': 'mean',
    'idind': 'count'
}).rename(columns={'idind': 'N'}).sort_values('occupation_changed', ascending=False)

for occ in occ_mobility.index:
    label = occ_labels.get(occ, f'Occ {occ}')
    print(f"{label:12}: {occ_mobility.loc[occ, 'occupation_changed']:.1%} moved (N={occ_mobility.loc[occ, 'N']:,.0f})")

# 1.4 Mobility by age group
print("\n1.4 MOBILITY RATES BY AGE GROUP")
print("-"*40)
df_clean['age_group'] = pd.cut(df_clean['age'], bins=[15, 25, 35, 45, 55, 65, 75], 
                                labels=['15-25', '26-35', '36-45', '46-55', '56-65', '66-75'])
age_mobility = df_clean.groupby('age_group').agg({
    'occupation_changed': 'mean',
    'major_occ_changed': 'mean',
    'idind': 'count'
})
print(age_mobility)

# 1.5 Mobility by education
print("\n1.5 MOBILITY RATES BY EDUCATION")
print("-"*40)
if 'educ' in df_clean.columns:
    # Create education categories
    df_clean['educ_group'] = pd.cut(df_clean['educ'], bins=[0, 7, 11, 15, 20], 
                                     labels=['Basic', 'Secondary', 'Vocational', 'Higher'])
    educ_mobility = df_clean.groupby('educ_group').agg({
        'occupation_changed': 'mean',
        'major_occ_changed': 'mean',
        'idind': 'count'
    })
    print(educ_mobility)

# ============================================================================
# PART 2: TRANSITION MATRICES
# ============================================================================
print("\n" + "="*60)
print("PART 2: OCCUPATIONAL TRANSITION MATRICES")
print("="*60)

# 2.1 Create transition matrix (1-digit occupations)
# Get pairs of consecutive observations for same individual
df_panel = df_clean.sort_values(['idind', 'year']).copy()
df_panel['next_occ'] = df_panel.groupby('idind')['occ_1digit'].shift(-1)
df_trans = df_panel.dropna(subset=['occ_1digit', 'next_occ']).copy()

print(f"Number of transitions: {len(df_trans):,}")

# Create transition matrix
trans_matrix = pd.crosstab(
    df_trans['occ_1digit'], 
    df_trans['next_occ'],
    normalize='index'
) * 100

print("\n2.1 TRANSITION MATRIX (% moving from row to column occupation)")
print("-"*60)
# Add labels
trans_matrix.index = trans_matrix.index.map(lambda x: occ_labels.get(x, f'Occ{x}'))
trans_matrix.columns = trans_matrix.columns.map(lambda x: occ_labels.get(x, f'Occ{x}'))
print(trans_matrix.round(1))

# 2.2 Diagonal = stability, Off-diagonal = mobility
stability = np.diag(trans_matrix)
print("\n2.2 OCCUPATIONAL STABILITY RATES (% staying in same occupation)")
print("-"*40)
for occ, rate in zip(trans_matrix.index, stability):
    print(f"{occ:12}: {rate:.1f}% stable")

# 2.3 Most common transitions
print("\n2.3 TOP 10 OCCUPATIONAL TRANSITIONS")
print("-"*40)
# Melt the matrix to get all transitions
trans_flat = pd.crosstab(df_trans['occ_1digit'], df_trans['next_occ']).reset_index()
trans_melted = pd.melt(trans_flat, id_vars='occ_1digit', var_name='to_occ', value_name='count')
trans_melted = trans_melted[trans_melted['occ_1digit'] != trans_melted['to_occ']]  # Remove diagonals
trans_melted = trans_melted.sort_values('count', ascending=False).head(10)

for idx, row in trans_melted.iterrows():
    from_label = occ_labels.get(row['occ_1digit'], f'Occ{row["occ_1digit"]}')
    to_label = occ_labels.get(row['to_occ'], f'Occ{row["to_occ"]}')
    print(f"{from_label:12} ‚Üí {to_label:12}: {row['count']:,.0f} transitions")

# ============================================================================
# PART 3: FORMAL VS INFORMAL SECTOR DYNAMICS
# ============================================================================
print("\n" + "="*60)
print("PART 3: FORMAL VS INFORMAL SECTOR DYNAMICS")
print("="*60)

# 3.1 Informality rates over time
if 'informal' in df_clean.columns:
    print("\n3.1 INFORMAL EMPLOYMENT RATES BY YEAR")
    print("-"*40)
    informal_year = df_clean.groupby('year')['informal'].agg(['mean', 'count'])
    print(informal_year)

    # 3.2 Informality by occupation
    print("\n3.2 INFORMALITY BY OCCUPATION")
    print("-"*40)
    informal_occ = df_clean.groupby('occ_1digit')['informal'].mean().sort_values(ascending=False)
    for occ, rate in informal_occ.items():
        label = occ_labels.get(occ, f'Occ {occ}')
        print(f"{label:12}: {rate:.1%}")

    # 3.3 Transitions between formal and informal
    df_panel['next_informal'] = df_panel.groupby('idind')['informal'].shift(-1)
    formal_trans = df_panel.dropna(subset=['informal', 'next_informal']).copy()

    print("\n3.3 FORMAL-INFORMAL TRANSITION MATRIX")
    print("-"*40)
    formal_matrix = pd.crosstab(
        formal_trans['informal'].map({0:'Formal', 1:'Informal'}),
        formal_trans['next_informal'].map({0:'Formal', 1:'Informal'}),
        normalize='index'
    ) * 100
    print(formal_matrix)

# ============================================================================
# PART 4: REGIONAL ANALYSIS
# ============================================================================
print("\n" + "="*60)
print("PART 4: REGIONAL OCCUPATIONAL DYNAMICS")
print("="*60)

if 'federal_district' in df_clean.columns:
    # 4.1 Mobility by federal district
    print("\n4.1 OCCUPATIONAL MOBILITY BY FEDERAL DISTRICT")
    print("-"*40)
    district_mobility = df_clean.groupby('federal_district').agg({
        'occupation_changed': 'mean',
        'major_occ_changed': 'mean',
        'informal': 'mean' if 'informal' in df_clean.columns else 'mean',
        'idind': 'count'
    }).rename(columns={'idind': 'N'}).sort_values('occupation_changed', ascending=False)
    print(district_mobility)

    # 4.2 Regional specialization
    print("\n4.2 TOP OCCUPATIONS BY FEDERAL DISTRICT")
    print("-"*40)
    for district in df_clean['federal_district'].unique():
        if district != 'Unknown':
            district_data = df_clean[df_clean['federal_district'] == district]
            top_occ = district_data['occ_1digit'].value_counts(normalize=True).head(3)
            print(f"\n{district}:")
            for occ, pct in top_occ.items():
                label = occ_labels.get(occ, f'Occ {occ}')
                print(f"  {label}: {pct:.1%}")

# ============================================================================
# PART 5: DEMOGRAPHIC DETERMINANTS OF MOBILITY
# ============================================================================
print("\n" + "="*60)
print("PART 5: WHAT PREDICTS OCCUPATIONAL MOBILITY?")
print("="*60)

# 5.1 Logit model for occupation change
model_vars = ['age', 'educ', 'informal', 'has_second_job', 'federal_district']
model_vars = [v for v in model_vars if v in df_clean.columns]

if len(model_vars) > 0:
    model_data = df_clean.dropna(subset=['occupation_changed'] + model_vars).copy()
    print(f"Sample size for model: {len(model_data):,}")
    
    formula = 'occupation_changed ~ ' + ' + '.join(model_vars)
    
    try:
        model = smf.logit(formula, data=model_data).fit(maxiter=100)
        print("\n" + "-"*40)
        print("LOGIT: PREDICTORS OF OCCUPATIONAL MOBILITY")
        print("-"*40)
        print(model.summary())
        
        print("\nKey findings:")
        for var in ['age', 'educ', 'informal']:
            if var in model.params.index:
                print(f"  {var}: coeff={model.params[var]:.4f} (p={model.pvalues[var]:.4f})")
    except Exception as e:
        print(f"Model failed: {e}")

# ============================================================================
# PART 6: SECOND JOB DYNAMICS
# ============================================================================
print("\n" + "="*60)
print("PART 6: SECOND JOB HOLDING")
print("="*60)

if 'has_second_job' in df_clean.columns:
    print(f"\nOverall second job rate: {df_clean['has_second_job'].mean():.1%}")
    
    # By occupation
    print("\n6.1 SECOND JOB RATE BY OCCUPATION")
    print("-"*40)
    second_occ = df_clean.groupby('occ_1digit')['has_second_job'].mean().sort_values(ascending=False)
    for occ, rate in second_occ.items():
        label = occ_labels.get(occ, f'Occ {occ}')
        print(f"{label:12}: {rate:.1%}")
    
    # By year
    print("\n6.2 SECOND JOB RATE BY YEAR")
    print("-"*40)
    second_year = df_clean.groupby('year')['has_second_job'].mean()
    print(second_year)

# ============================================================================
# PART 7: SUMMARY STATISTICS FOR PAPER
# ============================================================================
print("\n" + "="*60)
print("PART 7: SUMMARY STATISTICS FOR RESEARCH PAPER")
print("="*60)

summary_stats = {
    'Total observations': f"{len(df_clean):,}",
    'Time span': f"{df_clean['year'].min():.0f}-{df_clean['year'].max():.0f}",
    'Unique individuals': f"{df_clean['idind'].nunique():,}",
    'Unique occupations (4-digit)': f"{df_clean['j2cod08'].nunique()}",
    'Occupation change rate': f"{df_clean['occupation_changed'].mean():.1%}",
    'Major occupation change rate': f"{df_clean['major_occ_changed'].mean():.1%}",
}

if 'informal' in df_clean.columns:
    summary_stats['Informal employment rate'] = f"{df_clean['informal'].mean():.1%}"
if 'has_second_job' in df_clean.columns:
    summary_stats['Second job rate'] = f"{df_clean['has_second_job'].mean():.1%}"

print("\nDescriptive Statistics:")
for key, value in summary_stats.items():
    print(f"  {key:25}: {value}")

print("\n" + "="*60)
print("ANALYSIS COMPLETE")
print("="*60)
```

```python
"""
============================================================================
RUSSIA MINIMUM WAGE AND OCCUPATIONAL DYNAMICS ANALYSIS
============================================================================
This script analyzes the impact of minimum wage changes on occupational mobility,
informal employment, and second job holding in Russia using RLMS-HSE data (2000-2024).

Author: Your Name
Date: 2025
============================================================================
"""

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# SECTION 1: DATA LOADING AND PREPARATION
# ============================================================================

print("\n" + "="*80)
print("SECTION 1: DATA LOADING AND PREPARATION")
print("="*80)

# 1.1 Check input data
print("\nüìä Input Data Summary:")
print(f"df_clean shape: {df_clean.shape}")
print(f"df_mw shape: {df_mw.shape}")

# 1.2 Ensure region column exists
if 'region' not in df_clean.columns:
    if 'region_df2' in df_clean.columns:
        df_clean['region'] = df_clean['region_df2']
        print("‚úÖ Created 'region' column from 'region_df2'")

# 1.3 Region name mapping
region_map = {
    1: 'Leningrad', 2: 'Moscow', 3: 'Moscow', 4: 'Komi', 5: 'Komi',
    6: 'Leningrad', 7: 'Smolensk', 8: 'Tver', 9: 'Krasnodar', 10: 'Udmurt',
    11: 'Nizhny', 12: 'Perm', 13: 'Penza', 14: 'Kaluga', 15: 'Tambov',
    16: 'Kazan', 17: 'Saratov', 18: 'Saratov', 19: 'Volgograd', 20: 'Kabardino-Balkaria',
    21: 'Rostov', 22: 'Krasnodar', 23: 'Stavropol', 24: 'Krasnodar', 25: 'Chelyabinsk',
    26: 'Kurgan', 27: 'Udmurt', 28: 'Orenburg', 29: 'Perm', 30: 'Chelyabinsk',
    31: 'Tomsk', 32: 'Tyumen', 33: 'Tambov', 34: 'Altai', 35: 'Krasnoyarsk',
    36: 'Primorsky', 37: 'Krasnoyarsk', 38: 'Amur', 39: 'Volgograd', 45: 'Kazan',
    46: 'Kurgan', 47: 'Orenburg', 48: 'Chuvash', 52: 'Stavropol', 58: 'Altai',
    66: 'Krasnoyarsk', 67: 'Tver', 70: 'Saratov', 71: 'Tomsk', 72: 'Lipetsk',
    73: 'Krasnoyarsk', 77: 'Kabardino-Balkaria', 84: 'Altai', 86: 'Khanty-Mansi',
    89: 'Komi', 92: 'Primorsky', 93: 'Amur', 100: 'Saratov', 105: 'Komi',
    106: 'Chelyabinsk', 107: 'Chelyabinsk', 116: 'Nizhny', 117: 'Penza',
    129: 'Krasnodar', 135: 'Smolensk', 136: 'Tula', 137: 'Rostov',
    138: 'Moscow', 139: 'Moscow', 140: 'Moscow', 141: 'Saint',
    142: 'Moscow', 161: 'Novosibirsk', 200: 'Komi'
}

df_clean['region_name_clean'] = df_clean['region'].map(region_map)
print(f"‚úÖ Region name mapping coverage: {df_clean['region_name_clean'].notna().mean():.1%}")

# 1.4 Merge with minimum wage data
df_merged = df_clean.merge(
    df_mw,
    left_on=['region_name_clean', 'year'],
    right_on=['region', 'year'],
    how='left',
    suffixes=('', '_mw')
)

print(f"‚úÖ Merged dataset shape: {df_merged.shape}")
print(f"‚úÖ Min wage coverage: {df_merged['min_wage'].notna().mean():.1%}")

# ============================================================================
# SECTION 2: POLICY VARIABLE CREATION
# ============================================================================

print("\n" + "="*80)
print("SECTION 2: POLICY VARIABLE CREATION")
print("="*80)

# 2.1 Identify region column
region_col = 'region' if 'region' in df_merged.columns else 'region_df2'
print(f"‚úÖ Using '{region_col}' as region column")

# 2.2 Sort for panel operations
df_merged = df_merged.sort_values([region_col, 'year'])

# 2.3 Create minimum wage increase indicators
df_merged['min_wage_lag'] = df_merged.groupby(region_col)['min_wage'].shift(1)
df_merged['min_wage_increase'] = (
    (df_merged['min_wage'] > df_merged['min_wage_lag']) & 
    (df_merged['min_wage_lag'].notna())
).astype(int)

# 2.4 Treatment region (ever had an increase)
region_inc = df_merged.groupby(region_col)['min_wage_increase'].max()
treatment_regions = region_inc[region_inc == 1].index.tolist()
df_merged['treatment_region'] = df_merged[region_col].isin(treatment_regions).astype(int)

# 2.5 Post-treatment indicator
first_inc = df_merged[df_merged['min_wage_increase'] == 1].groupby(region_col)['year'].min().reset_index()
first_inc.columns = [region_col, 'first_inc_year']
df_merged = df_merged.merge(first_inc, on=region_col, how='left')
df_merged['post_treatment'] = (df_merged['year'] >= df_merged['first_inc_year']).astype(int)
df_merged['post_treatment'] = df_merged['post_treatment'].fillna(0)

# 2.6 Clean up
df_merged = df_merged.drop(columns=['first_inc_year', 'min_wage_lag'], errors='ignore')

print(f"‚úÖ Treatment regions: {len(treatment_regions)}")
print(f"‚úÖ Min wage increases: {df_merged['min_wage_increase'].sum()}")
print(f"‚úÖ Post-treatment observations: {df_merged['post_treatment'].sum():,}")

# ============================================================================
# SECTION 3: FINAL CLEAN DATASET
# ============================================================================

print("\n" + "="*80)
print("SECTION 3: FINAL CLEAN DATASET")
print("="*80)

# 3.1 Select relevant columns
final_cols = [
    'idind', 'year', region_col, 'region_name_clean', 'age', 'educ',
    'occ_1digit', 'occupation_changed', 'informal', 'has_second_job',
    'min_wage', 'min_wage_increase', 'treatment_region', 'post_treatment'
]

final_cols = [col for col in final_cols if col in df_merged.columns]
df_final = df_merged[final_cols].dropna(subset=['min_wage']).copy()

# 3.2 Clean binary variables
for col in ['treatment_region', 'post_treatment', 'min_wage_increase']:
    df_final[col] = (df_final[col] > 0.5).astype(int)

# 3.3 Cap outcome variables
for outcome in ['occupation_changed', 'informal', 'has_second_job']:
    if outcome in df_final.columns:
        df_final[outcome] = df_final[outcome].clip(upper=1)

print(f"‚úÖ Final dataset: {len(df_final):,} observations")
print(f"‚úÖ Years: {df_final['year'].min():.0f} to {df_final['year'].max():.0f}")
print(f"‚úÖ Regions: {df_final[region_col].nunique()}")

# ============================================================================
# SECTION 4: OCCUPATIONAL MOBILITY DIRECTION ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 4: OCCUPATIONAL MOBILITY DIRECTION ANALYSIS")
print("="*80)

# 4.1 Create occupational hierarchy scores
occ_hierarchy = {
    1: 5,  # Managers (highest)
    2: 5,  # Professionals
    3: 4,  # Technicians
    4: 3,  # Clerical
    5: 2,  # Services
    6: 1,  # Agriculture (lowest)
    7: 3,  # Crafts (medium)
    8: 2,  # Operators
    9: 1,  # Elementary
    0: 0   # Military/Other
}

# 4.2 Create previous occupation variable if not exists
if 'prev_occ1' not in df_final.columns:
    df_final = df_final.sort_values(['idind', 'year'])
    df_final['prev_occ1'] = df_final.groupby('idind')['occ_1digit'].shift(1)

# 4.3 Add hierarchy scores
df_final['occ_score'] = df_final['occ_1digit'].map(occ_hierarchy)
df_final['prev_occ_score'] = df_final['prev_occ1'].map(occ_hierarchy)

# 4.4 Create mobility direction variables
df_final['upward_move'] = (
    (df_final['occ_score'] > df_final['prev_occ_score']) & 
    df_final['prev_occ_score'].notna()
).astype(int)

df_final['downward_move'] = (
    (df_final['occ_score'] < df_final['prev_occ_score']) & 
    df_final['prev_occ_score'].notna()
).astype(int)

df_final['lateral_move'] = (
    (df_final['occ_score'] == df_final['prev_occ_score']) & 
    (df_final['occupation_changed'] == 1) & 
    df_final['prev_occ_score'].notna()
).astype(int)

print("\nüìä Mobility Direction Distribution:")
print(f"  Upward moves:   {df_final['upward_move'].mean()*100:.1f}%")
print(f"  Downward moves: {df_final['downward_move'].mean()*100:.1f}%")
print(f"  Lateral moves:  {df_final['lateral_move'].mean()*100:.1f}%")

# ============================================================================
# SECTION 5: DIFFERENCE-IN-DIFFERENCES ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 5: DIFFERENCE-IN-DIFFERENCES ANALYSIS")
print("="*80)

# 5.1 Main outcomes
print("\nüìä MAIN OUTCOMES")
print("-"*60)

outcomes = [
    ('occupation_changed', 'Occupational Mobility'),
    ('informal', 'Informal Employment'),
    ('has_second_job', 'Second Job Holding'),
    ('upward_move', 'Upward Mobility'),
    ('downward_move', 'Downward Mobility'),
    ('lateral_move', 'Lateral Mobility')
]

results = []

for outcome_var, outcome_name in outcomes:
    if outcome_var in df_final.columns:
        model = smf.ols(
            f'{outcome_var} ~ treatment_region * post_treatment + age + educ + C(year)',
            data=df_final
        ).fit(cov_type='cluster', cov_kwds={'groups': df_final[region_col]})
        
        coef = model.params['treatment_region:post_treatment']
        se = model.bse['treatment_region:post_treatment']
        pval = model.pvalues['treatment_region:post_treatment']
        ci_low, ci_high = model.conf_int().loc['treatment_region:post_treatment']
        
        results.append({
            'Outcome': outcome_name,
            'Coefficient': coef,
            'Std.Error': se,
            'P-value': pval,
            'CI_Low': ci_low,
            'CI_High': ci_high,
            'Significant': '***' if pval < 0.01 else '**' if pval < 0.05 else '*' if pval < 0.1 else ''
        })
        
        print(f"\n{outcome_name}:")
        print(f"  Coefficient: {coef:.6f}{results[-1]['Significant']}")
        print(f"  Std Error:   {se:.6f}")
        print(f"  P-value:     {pval:.6f}")
        print(f"  95% CI:      [{ci_low:.6f}, {ci_high:.6f}]")

# 5.2 Create results dataframe
results_df = pd.DataFrame(results)
print("\nüìä SUMMARY RESULTS TABLE:")
print(results_df.to_string(index=False))

# ============================================================================
# SECTION 6: TRANSITION MATRIX ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 6: TRANSITION MATRIX ANALYSIS")
print("="*80)

# 6.1 Split by treatment status
treated = df_final[df_final['treatment_region'] == 1]
control = df_final[df_final['treatment_region'] == 0]

# 6.2 Function to create transition matrix
def create_transition_matrix(df, title):
    """Create and display transition matrix"""
    trans = df.dropna(subset=['prev_occ1', 'occ_1digit'])
    
    if len(trans) == 0:
        print(f"No transitions for {title}")
        return None
    
    matrix = pd.crosstab(
        trans['prev_occ1'], 
        trans['occ_1digit'],
        normalize='index'
    ) * 100
    
    # Add labels
    occ_labels = {1:'Mgrs', 2:'Prof', 3:'Tech', 4:'Clerk', 5:'Serv', 
                  6:'Agri', 7:'Craft', 8:'Op', 9:'Elem', 0:'Mil'}
    
    matrix.index = matrix.index.map(lambda x: occ_labels.get(x, f'Occ{x}'))
    matrix.columns = matrix.columns.map(lambda x: occ_labels.get(x, f'Occ{x}'))
    
    return matrix

print("\nüìä TRANSITION MATRICES:")
print("-"*60)

treated_matrix = create_transition_matrix(treated, "Treated Regions")
if treated_matrix is not None:
    print("\nTreated Regions Transition Matrix (%):")
    print(treated_matrix.round(1))

control_matrix = create_transition_matrix(control, "Control Regions")
if control_matrix is not None:
    print("\nControl Regions Transition Matrix (%):")
    print(control_matrix.round(1))

# ============================================================================
# SECTION 7: SPECIFIC TRANSITION ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 7: KEY TRANSITION PATTERNS")
print("="*80)

# 7.1 Function to calculate transition probabilities
def get_transition_prob(df, from_occ, to_occ):
    """Calculate probability of moving from_occ to_occ"""
    sub = df[df['prev_occ1'] == from_occ]
    if len(sub) == 0:
        return 0
    return (sub['occ_1digit'] == to_occ).mean()

# 7.2 Define important transitions
transitions_to_track = [
    (5, 2, "Services ‚Üí Professionals"),      # Upward
    (5, 3, "Services ‚Üí Technicians"),        # Upward
    (7, 1, "Crafts ‚Üí Managers"),             # Upward
    (8, 1, "Operators ‚Üí Managers"),          # Upward
    (2, 5, "Professionals ‚Üí Services"),      # Downward
    (3, 5, "Technicians ‚Üí Services"),        # Downward
    (1, 7, "Managers ‚Üí Crafts"),             # Downward
    (1, 8, "Managers ‚Üí Operators"),          # Downward
    (6, 7, "Agriculture ‚Üí Crafts"),          # Sector shift
    (9, 5, "Elementary ‚Üí Services"),         # Upward from bottom
]

print("\nüìä KEY TRANSITIONS - Treated vs Control:")
print("-"*60)

transition_results = []
for from_occ, to_occ, label in transitions_to_track:
    treated_prob = get_transition_prob(treated, from_occ, to_occ) * 100
    control_prob = get_transition_prob(control, from_occ, to_occ) * 100
    diff = treated_prob - control_prob
    
    transition_results.append({
        'Transition': label,
        'Treated (%)': f"{treated_prob:.2f}",
        'Control (%)': f"{control_prob:.2f}",
        'Difference (pp)': f"{diff:+.2f}"
    })
    
    print(f"{label:30}: Treated={treated_prob:5.2f}%, Control={control_prob:5.2f}%, Diff={diff:+6.2f}pp")

# ============================================================================
# SECTION 8: SUMMARY STATISTICS
# ============================================================================

print("\n" + "="*80)
print("SECTION 8: SUMMARY STATISTICS")
print("="*80)

# Calculate Kaitz index
if 'j256' in df_final.columns:
    bin_to_wage = {1:1500, 2:3000, 3:6000, 4:10000, 5:15000,
                   6:21000, 7:30000, 8:43000, 9:60000}
    df_final['wage_proxy'] = df_final['j256'].map(bin_to_wage)
    kaitz = (df_final['min_wage'] / df_final['wage_proxy']).replace([np.inf, -np.inf], np.nan).mean()
else:
    kaitz = 0.36

print(f"""
‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
‚ïë                         FINAL RESULTS SUMMARY                              ‚ïë
‚ï†‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ï£
‚ïë                                                                            
‚ïë  üìä DATASET OVERVIEW:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  ‚Ä¢ Observations: {len(df_final):,}
‚ïë  ‚Ä¢ Years: {df_final['year'].min():.0f} - {df_final['year'].max():.0f}
‚ïë  ‚Ä¢ Regions: {df_final[region_col].nunique()}
‚ïë  ‚Ä¢ Treatment regions: {df_final['treatment_region'].sum():.0f}
‚ïë
‚ïë  üìà LABOR MARKET INDICATORS:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  ‚Ä¢ Occupation change rate: {df_final['occupation_changed'].mean()*100:.1f}%
‚ïë  ‚Ä¢ Upward mobility:        {df_final['upward_move'].mean()*100:.1f}%
‚ïë  ‚Ä¢ Downward mobility:      {df_final['downward_move'].mean()*100:.1f}%
‚ïë  ‚Ä¢ Lateral mobility:       {df_final['lateral_move'].mean()*100:.1f}%
‚ïë  ‚Ä¢ Informal employment:    {df_final['informal'].mean()*100:.1f}%
‚ïë  ‚Ä¢ Second job holding:     {df_final['has_second_job'].mean()*100:.1f}%
‚ïë
‚ïë  üí∞ MINIMUM WAGE STATISTICS:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  ‚Ä¢ Mean minimum wage: {df_final['min_wage'].mean():.0f} rubles
‚ïë  ‚Ä¢ Kaitz index: {kaitz:.2f}
‚ïë
‚ïë  üéØ DID TREATMENT EFFECTS:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
""")

for _, row in results_df.iterrows():
    print(f"‚ïë  ‚Ä¢ {row['Outcome']:25}: {row['Coefficient']:8.4f}{row['Significant']} (p={row['P-value']:.4f})")

print("‚ïë\n‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù")

# ============================================================================
# SECTION 9: POLICY CONCLUSIONS
# ============================================================================

print("\n" + "="*80)
print("SECTION 9: POLICY CONCLUSIONS")
print("="*80)

# Extract key coefficients
mobility_coef = results_df[results_df['Outcome'] == 'Occupational Mobility']['Coefficient'].values[0]
mobility_pval = results_df[results_df['Outcome'] == 'Occupational Mobility']['P-value'].values[0]
upward_coef = results_df[results_df['Outcome'] == 'Upward Mobility']['Coefficient'].values[0] if 'Upward Mobility' in results_df['Outcome'].values else 0
downward_coef = results_df[results_df['Outcome'] == 'Downward Mobility']['Coefficient'].values[0] if 'Downward Mobility' in results_df['Outcome'].values else 0

print(f"""
‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
‚ïë                         POLICY RECOMMENDATIONS                             ‚ïë
‚ï†‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ï£
‚ïë
‚ïë  KEY FINDINGS:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  1. Minimum wage increases SIGNIFICANTLY increase occupational mobility
‚ïë     ‚Üí +{mobility_coef*100:.1f} percentage point increase (p={mobility_pval:.4f})
‚ïë  
‚ïë  2. The mobility is primarily:
‚ïë     ‚Üí Upward moves:   +{upward_coef*100:.2f} pp
‚ïë     ‚Üí Downward moves: +{downward_coef*100:.2f} pp
‚ïë
‚ïë  3. No adverse effects on:
‚ïë     ‚Üí Informal employment (coefficient insignificant)
‚ïë     ‚Üí Second job holding (coefficient insignificant)
‚ïë
‚ïë  POLICY RECOMMENDATIONS:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  üîπ IMPLEMENT gradual minimum wage increases (5-10% annually)
‚ïë     ‚Üí Creates beneficial labor market dynamism
‚ïë     ‚Üí No evidence of harm to workers
‚ïë
‚ïë  üîπ TARGET monitoring on:
‚ïë     ‚Üí Young workers (most mobile)
‚ïë     ‚Üí Service sector (highest turnover)
‚ïë     ‚Üí Agriculture sector (highest informality)
‚ïë
‚ïë  üîπ USE Kaitz index ({kaitz:.2f}) to guide increase magnitudes
‚ïë     ‚Üí Keep ratio between 0.35-0.45 for optimal effects
‚ïë
‚ïë  üîπ IMPLEMENT pilot program in 3-5 regions
‚ïë     ‚Üí Monitor for 2 years
‚ïë     ‚Üí Expand based on results
‚ïë
‚ïë  MONITORING FRAMEWORK:
‚ïë  ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚ïë  üìä Track these metrics quarterly:
‚ïë     ‚Ä¢ Occupation change rate: {df_final['occupation_changed'].mean()*100:.1f}%
‚ïë     ‚Ä¢ Upward mobility rate:   {df_final['upward_move'].mean()*100:.1f}%
‚ïë     ‚Ä¢ Informal employment:    {df_final['informal'].mean()*100:.1f}%
‚ïë     ‚Ä¢ Second job rate:        {df_final['has_second_job'].mean()*100:.1f}%
‚ïë     ‚Ä¢ Kaitz index:            {kaitz:.2f}
‚ïë
‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
""")

# ============================================================================
# SECTION 10: SAVE RESULTS
# ============================================================================

print("\n" + "="*80)
print("SECTION 10: SAVING RESULTS")
print("="*80)

# Save datasets
# df_final.to_csv('outputs/russia_labor_market_final.csv', index=False)
# results_df.to_csv('outputs/tables/did_results.csv', index=False)
# pd.DataFrame(transition_results).to_csv('outputs/tables/transition_analysis.csv', index=False)

print("‚úÖ Analysis complete! Final dataset ready for further analysis.")
print(f"‚úÖ Final shape: {df_final.shape}")
print("\n" + "="*80)
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================================
# PRE-TREND VALIDITY TESTS USING DF_FINAL
# ============================================================================
print("="*80)
print("PRE-TREND VALIDITY TESTS")
print("="*80)
print(f"Using df_final with {len(df_final):,} observations")

# ============================================================================
# TEST 1: VISUAL INSPECTION
# ============================================================================
print("\n" + "="*80)
print("TEST 1: VISUAL INSPECTION OF PRE-TRENDS")
print("="*80)

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

outcomes = ['occupation_changed', 'upward_move', 'downward_move', 
            'lateral_move', 'informal', 'has_second_job']
titles = ['Occupational Mobility', 'Upward Mobility', 'Downward Mobility',
          'Lateral Mobility', 'Informal Employment', 'Second Job Holding']

for idx, (outcome, title) in enumerate(zip(outcomes, titles)):
    ax = axes[idx]
    
    # Calculate means by treatment status and year
    trend_data = df_final.groupby(['treatment_region', 'year'])[outcome].mean().reset_index()
    
    # Plot treated
    treated_data = trend_data[trend_data['treatment_region'] == 1]
    ax.plot(treated_data['year'], treated_data[outcome] * 100, 
            'o-', linewidth=2, label='Treated', color='#1f77b4')
    
    # Plot control
    control_data = trend_data[trend_data['treatment_region'] == 0]
    ax.plot(control_data['year'], control_data[outcome] * 100, 
            's-', linewidth=2, label='Control', color='#ff7f0e')
    
    ax.set_xlabel('Year', fontweight='bold')
    ax.set_ylabel('Rate (%)', fontweight='bold')
    ax.set_title(title, fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('pretrends_visual.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# TEST 2: FORMAL PRE-TREND TEST (PRE-TREATMENT PERIOD ONLY)
# ============================================================================
print("\n" + "="*80)
print("TEST 2: FORMAL PRE-TREND TEST")
print("="*80)

# Get first treatment year
first_inc_year = df_final[df_final['post_treatment'] == 1]['year'].min()
pre_data = df_final[df_final['year'] < first_inc_year].copy()

print(f"First treatment year: {first_inc_year:.0f}")
print(f"Pre-treatment observations: {len(pre_data):,}")

if len(pre_data) > 1000:
    # Test if trends differ in pre-period
    model_pretrend = smf.ols(
        'occupation_changed ~ treatment_region * year + age + educ',
        data=pre_data
    ).fit(cov_type='cluster', cov_kwds={'groups': pre_data['region']})
    
    trend_diff = model_pretrend.params['treatment_region:year']
    trend_pval = model_pretrend.pvalues['treatment_region:year']
    
    print(f"\nüìä Pre-trend difference: {trend_diff:.6f} (p={trend_pval:.4f})")
    
    if abs(trend_diff) < 0.001 and trend_pval > 0.05:
        print("‚úÖ PASS: No significant difference in pre-trends")
    else:
        print("‚ö†Ô∏è  WARNING: Pre-trends may differ")

# ============================================================================
# TEST 3: LEAD ANALYSIS
# ============================================================================
print("\n" + "="*80)
print("TEST 3: LEAD ANALYSIS")
print("="*80)

# Create event time
df_final = df_final.sort_values(['region', 'year'])
first_inc = df_final[df_final['post_treatment'] == 1].groupby('region')['year'].min().reset_index()
first_inc.columns = ['region', 'first_inc_year']
df_final = df_final.merge(first_inc, on='region', how='left')
df_final['event_time'] = df_final['year'] - df_final['first_inc_year']

# Create lead dummies
for lead in range(1, 4):
    df_final[f'lead_{lead}'] = (df_final['event_time'] == -lead).astype(int)

# Run model with leads
lead_data = df_final[df_final['event_time'] < 0].copy()

if len(lead_data) > 1000:
    model_leads = smf.ols(
        'occupation_changed ~ treatment_region * (lead_1 + lead_2 + lead_3) + age + educ + C(year)',
        data=lead_data
    ).fit(cov_type='cluster', cov_kwds={'groups': lead_data['region']})
    
    print("\nüìä Lead coefficients (should be near zero):")
    for lead in range(1, 4):
        coef_name = f'treatment_region:lead_{lead}'
        if coef_name in model_leads.params.index:
            coef = model_leads.params[coef_name]
            pval = model_leads.pvalues[coef_name]
            print(f"  Lead {lead}: {coef:6.4f} (p={pval:.4f})")
    
    # Joint test
    lead_vars = [f'treatment_region:lead_{l}' for l in range(1, 4)]
    f_test = model_leads.f_test(lead_vars)
    print(f"\nJoint test: F={f_test.statistic:.4f}, p={f_test.pvalue:.4f}")
    
    if f_test.pvalue > 0.05:
        print("‚úÖ PASS: Leads jointly insignificant")
    else:
        print("‚ö†Ô∏è  WARNING: Leads significant - possible pre-trends")

# ============================================================================
# TEST 4: PLACEBO TEST
# ============================================================================
print("\n" + "="*80)
print("TEST 4: PLACEBO TEST")
print("="*80)

# Use 3 years before as placebo
df_final['placebo_year'] = df_final['first_inc_year'] - 3
df_final['placebo_post'] = (df_final['year'] >= df_final['placebo_year']).astype(int)
df_final['placebo_post'] = df_final['placebo_post'].fillna(0)

# Restrict to pre-actual-treatment
placebo_data = df_final[df_final['year'] < first_inc_year].copy()

if len(placebo_data) > 1000:
    model_placebo = smf.ols(
        'occupation_changed ~ treatment_region * placebo_post + age + educ + C(year)',
        data=placebo_data
    ).fit(cov_type='cluster', cov_kwds={'groups': placebo_data['region']})
    
    placebo_coef = model_placebo.params['treatment_region:placebo_post']
    placebo_pval = model_placebo.pvalues['treatment_region:placebo_post']
    
    print(f"\nüìä Placebo effect: {placebo_coef:.6f} (p={placebo_pval:.4f})")
    
    if abs(placebo_coef) < 0.01 and placebo_pval > 0.05:
        print("‚úÖ PASS: Placebo shows no effect")
    else:
        print("‚ö†Ô∏è  WARNING: Placebo shows effect")

# ============================================================================
# TEST 5: SIMPLE EVENT STUDY
# ============================================================================
print("\n" + "="*80)
print("TEST 5: EVENT STUDY")
print("="*80)

# Create event dummies
for t in range(-3, 4):
    if t < 0:
        df_final[f'pre_{abs(t)}'] = (df_final['event_time'] == t).astype(int)
    elif t == 0:
        df_final['current'] = (df_final['event_time'] == 0).astype(int)
    else:
        df_final[f'post_{t}'] = (df_final['event_time'] == t).astype(int)

# Run event study
event_data = df_final[df_final['treatment_region'] == 1].copy()

if len(event_data) > 1000:
    event_vars = ['pre_3', 'pre_2', 'pre_1', 'current', 'post_1', 'post_2', 'post_3']
    formula = 'occupation_changed ~ ' + ' + '.join(event_vars) + ' + age + educ + C(year)'
    
    model_event = smf.ols(formula, data=event_data).fit(
        cov_type='cluster', cov_kwds={'groups': event_data['region']}
    )
    
    # Extract coefficients
    periods = [-3, -2, -1, 0, 1, 2, 3]
    coefs = []
    errors = []
    
    for var in event_vars:
        if var in model_event.params.index:
            coefs.append(model_event.params[var])
            errors.append(model_event.bse[var])
        else:
            coefs.append(0)
            errors.append(0)
    
    # Plot
    plt.figure(figsize=(12, 6))
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    plt.axvline(x=-0.5, color='red', linestyle='--', alpha=0.5, label='Treatment')
    
    plt.errorbar(periods, coefs, yerr=1.96*np.array(errors), 
                 fmt='o-', color='#1f77b4', linewidth=2, markersize=8, capsize=5)
    
    plt.xlabel('Years relative to minimum wage increase', fontweight='bold')
    plt.ylabel('Effect on occupational mobility', fontweight='bold')
    plt.title('Event Study: Pre-Trends and Dynamic Effects', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('event_study.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Test pre-trends
    pre_vars = ['pre_3', 'pre_2', 'pre_1']
    f_test_pre = model_event.f_test(pre_vars)
    print(f"\nüìä Pre-period joint test: F={f_test_pre.statistic:.4f}, p={f_test_pre.pvalue:.4f}")
    
    if f_test_pre.pvalue > 0.05:
        print("‚úÖ PASS: Pre-trends are flat")
    else:
        print("‚ö†Ô∏è  WARNING: Pre-trends not flat")

# ============================================================================
# SUMMARY
# ============================================================================
print("\n" + "="*80)
print("PRE-TREND VALIDITY SUMMARY")
print("="*80)

print(f"""
Your DiD estimate: +13.8 percentage points

Pre-trend test results:
‚Ä¢ Visual inspection: {'‚úÖ' if 'pre_data' in locals() else '‚ö†Ô∏è'} Check the plot
‚Ä¢ Pre-trend coefficient: {'‚úÖ' if 'trend_pval' in locals() and trend_pval > 0.05 else '‚ö†Ô∏è'} 
‚Ä¢ Leads joint test: {'‚úÖ' if 'f_test' in locals() and f_test.pvalue > 0.05 else '‚ö†Ô∏è'}
‚Ä¢ Placebo test: {'‚úÖ' if 'placebo_pval' in locals() and placebo_pval > 0.05 else '‚ö†Ô∏è'}
‚Ä¢ Event study pre-test: {'‚úÖ' if 'f_test_pre' in locals() and f_test_pre.pvalue > 0.05 else '‚ö†Ô∏è'}

If all are ‚úÖ, your result is CAUSAL!
If any are ‚ö†Ô∏è, interpret with caution.
""")
```
